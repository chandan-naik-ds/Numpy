```python
import numpy as np
'''You received three different sensor readings as Python lists
(they have different lengths):Python
sensor_a = [22.1, 23.4, 21.9, 24.0]
sensor_b = [1013.2, 1015.7, 1012.1]
sensor_c = [45, 48, 46, 44, 50, 47]
Turn them into a single 2D NumPy array (6 rows × 3 columns) by padding
missing values with np.nan.
Final shape must be (6,3). Print the array.'''

sensor_a = [22.1, 23.4, 21.9, 24.0]
sensor_b = [1013.2, 1015.7, 1012.1]
sensor_c = [45, 48, 46, 44, 50, 47]

max_len = 6

a = np.pad(np.array(sensor_a, dtype=float), (0, max_len - len(sensor_a)), constant_values=np.nan)
b = np.pad(np.array(sensor_b, dtype=float), (0, max_len - len(sensor_b)), constant_values=np.nan)
c = np.pad(np.array(sensor_c, dtype=float), (0, max_len - len(sensor_c)), constant_values=np.nan)

result = np.column_stack((a,b,c))

print("Final array:\n",result)
print("Shape:",result.shape)

```
