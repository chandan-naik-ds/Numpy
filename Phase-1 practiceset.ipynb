```python
'''Element-Wise Ops: Create a list lst = [1, 2, 3, 4] and array arr = np.array([1, 2, 3, 4]). 
Add 10 to every element in both (loop for list, direct for array).
Time both ops with 10,000 repeats. Which is faster? Why? 
(Bonus: Try multiplying by 2.)'''

import numpy as np
import time 

list = [1, 2, 3, 4]
arr = np.array([1, 2, 3, 4])

#  -----list operation----

start = time.time()
for a in range(10000):
    result_list = []
    for i in list:
        result_list.append(i + 10)

end = time.time()
list_time = end - start


# -----Numpy operation-----
start = time.time()
for a in range(10000):
    result_arr = arr + 10

end = time.time()
arr_time = end - start


print("List result:",result_list)
print("List array:",result_arr)

print("List time: ", list_time)
print("Array time: ", arr_time)

# ----- Bonus Multiplication-----
start = time.time()
for a in range(10000):
    result_list_mul= []
    for i in list:
        result_list_mul.append(i * 2)
end = time.time()
print("List multiply time: ",end - start)

start = time.time()
for a in range(10000):
    result_arr_mul = arr * 2

end = time.time()
print("Array multiply time: ",end - start)

```

### Question 5

```python
'''Uniform Arrays: Create a 3x3 array of zeros (float dtype). Then, one of ones (int dtype). 
Stack them vertically with np.vstack and print the result. What's the new shape?'''

# zeros array
Zeros_array = np.zeros((3,3), dtype=float)

# ones array
Ones_array = np.ones((3,3), dtype=int)

# Verticle stack
result = np.vstack((Zeros_array,Ones_array))

print("Zeros array:\n",Zeros_array)
print("Oness array:\n",Ones_array)
print("Verticle array:\n",result)

print("Shape of reultant array:",result.shape)
```

```python
'''Sequence Generators: Use np.arange(0, 20, 2) to make an even-number array up to 18.
 Then np.linspace(0, 10, 6) for 6 evenly spaced points.
 Concat them horizontally (np.hstack)—what's the output?'''

even_array = np.arange(0, 20, 2)

linspace = np.linspace(0, 10, 6)

# Concating the both
result = np.hstack((even_array,linspace))


print("Even number array up to 18:\n",even_array)
print("Linspacee for 6 evenly spaced points:\n",linspace)
print("Concated array", result)
```

```python
'''Property Probe: Create a 2x4 array of random floats. Print its shape, ndim, size, and dtype. 
Flatten it (flatten()) vs. ravel (ravel())—are they the same? Modify one; does it affect the original?
(Key diff: copy vs. view.)'''

arr = np.random.rand(2,4)
print("Array: \n",arr)

# Properties
print("Shape: ", arr.shape)
print("Size: ", arr.size)
print("Dimension: ", arr.ndim)
print("Data type: ", arr.dtype)

# Flattenin the array  (CReates copy)
flat = arr.flatten()

# ravel (creates view)
raveled = arr.ravel()

print("\nFlattend:\n",flat)
print("Raveled:\n",raveled)

# flatten modification
flat[0] = 888

# raveled modification
raveled[1] = 999

print("\---nAfter modification---")
print("Flatten:\n",flat)
print("Raveled:\n",raveled)


print("\n---Original array---")
print(arr)
```
