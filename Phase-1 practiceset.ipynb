```python
'''Nested List to 2D Array: Turn [[1, 2, 3], [4, 5, 6]] into a NumPy array. Print its shape (arr.shape). 
What happens if you add a mismatched inner list like [7, 8]?'''

list1 = [[1, 2, 3], [4, 5, 6]]

arr = np.array(list1)
print(arr)
print("Dimension:",arr.ndim)
print("Shape:",arr.shape)

print("\n After adding mismatched list")
lst2 = [[1, 2, 3], [4, 5, 6], [7, 8]]
try :
    arr2 = np.array(lst2)
    print(arr2)
    print("Shape:", arr2.shape)
except ValueError as e:
        print("Error:", e)
```

### QUESTION---3

```python
'''Element-Wise Ops: Create a list lst = [1, 2, 3, 4] and array arr = np.array([1, 2, 3, 4]). 
Add 10 to every element in both (loop for list, direct for array).
Time both ops with 10,000 repeats. Which is faster? Why? 
(Bonus: Try multiplying by 2.)'''

import numpy as np
import time 

list = [1, 2, 3, 4]
arr = np.array([1, 2, 3, 4])

#  -----list operation----

start = time.time()
for a in range(10000):
    result_list = []
    for i in list:
        result_list.append(i + 10)

end = time.time()
list_time = end - start


# -----Numpy operation-----
start = time.time()
for a in range(10000):
    result_arr = arr + 10

end = time.time()
arr_time = end - start


print("List result:",result_list)
print("List array:",result_arr)

print("List time: ", list_time)
print("Array time: ", arr_time)

# ----- Bonus Multiplication-----
start = time.time()
for a in range(10000):
    result_list_mul= []
    for i in list:
        result_list_mul.append(i * 2)
end = time.time()
print("List multiply time: ",end - start)

start = time.time()
for a in range(10000):
    result_arr_mul = arr * 2

end = time.time()
print("Array multiply time: ",end - start)

```

```python
'''Slicing Surprise: Slice the first two elements from lst and arr above. Modify the slice (e.g., set to 99). Does it 
affect the original? Explain the mutability difference.'''

import numpy as np

lst = [1, 2, 3, 4]
arr = np.array([1, 2, 3, 4])

# Create slices
list_slice = lst[:2]
arr_slice = arr[:2]

print("List slice:", list_slice)
print("Array slice:", arr_slice)

# Modify slices
list_slice[0] = 99
arr_slice[0] = 99

print("\nModified slices:")
print("List slice:", list_slice)
print("Array slice:", arr_slice)

print("\nOriginal structures:")
print("Original list:", lst)
print("Original array:", arr)



```
