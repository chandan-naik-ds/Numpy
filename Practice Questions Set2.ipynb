13: Extract Odd Rows and Even Columns

```python
import numpy as np
sampleArray = np.array([
    [3, 6, 9, 12], 
    [15, 18, 21, 24], 
    [27, 30, 33, 36], 
    [39, 42, 45, 48],
    [51, 54, 57, 60]
])

new_array = sampleArray[1::2, ::2]

print("Original Array: \n",sampleArray)

print("New array with odd rows and even columns: \n",new_array)
```

14: Stack arrays horizontally

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

stacked_arr = np.hstack((a,b))
print("Horizontally Stacked Array: \n",stacked_arr)
```

 15: Slice the first two rows and first two columns from a 4×4 array

```python
arr = np.arange(1, 17).reshape(4,4)
print("Original AArray: \n",arr)

sliced_arr = arr[:2, :2]
print(sliced_arr)
```

16: Replace all odd numbers in a NumPy array with -1

```python
arr = np.arange(1, 11)
print("Original array: ",arr)

arr[arr % 2 == 1] = -1
print("modified_arr: ",arr)
```

17: Get the indices of non-zero elements in an array


```python
arr = np.array([1, 0, 2, 0, 3, 0, 4])
print("Array: ",arr)

indices = np.nonzero(arr)
print("Indices of non zero elements: ",indices)
```
