```python
arr = np.array([10, 20, 30, 100, 200, 300])

mean = np.mean(arr)
print("Mean: ",mean)

# or

mean = np.sum(arr)/len(arr)
print("Mean: ",mean)



median = np.median(arr)
print("Median: ",median)



variance = np.sum((arr-mean)**2)/ len(arr)
print("Variance: ",variance)

std_div = np.std(arr)
print("Standarad Deviation: ",std_div)

# or 

standard_deviation = np.sqrt(variance)
print(standard_deviation)
```

 22: Remove common items from array

```python


import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 6, 7, 8, 9])

# Create a boolean mask where `a`'s elements are NOT in `b`
mask = np.isin(a, b)
print(mask)

# Use the `~` operator to invert the mask (get the opposite)
result = b[~mask]
print(result)
```

 48: Find nearest value

Exercise 49: Convert to object array
Convert a 1D array to an object array, retaining the original data type of each element.

```python
arr = np.array([1, 2, 3, 4, 5])

obj_arr = np.empty(arr.size, dtype='object')

obj_arr[:] = arr

obj_arr[1] = 'a'

print(obj_arr)
print(obj_arr.dtype)
```
