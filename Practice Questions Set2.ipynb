21: Compute the mean, median, and standard deviation of a NumPy array

```python
arr = np.array([10, 20, 30, 100, 200, 300])

mean = np.mean(arr)
print("Mean: ",mean)

# or

mean = np.sum(arr)/len(arr)
print("Mean: ",mean)



median = np.median(arr)
print("Median: ",median)



variance = np.sum((arr-mean)**2)/ len(arr)
print("Variance: ",variance)

std_div = np.std(arr)
print("Standarad Deviation: ",std_div)

# or 

standard_deviation = np.sqrt(variance)
print(standard_deviation)
```

 22: Remove common items from array

```python


import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 6, 7, 8, 9])

# Create a boolean mask where `a`'s elements are NOT in `b`
mask = np.isin(a, b)
print(mask)

# Use the `~` operator to invert the mask (get the opposite)
result = b[~mask]
print(result)
```

23: Normalize a NumPy array (values between 0 and 1)

norm =   x - xmin  / xmax- xmin

```python
arr = np.array([10, 20, 30, 40, 50])

normailzed = (arr - arr.min())/ (arr.max() - arr.min())

print("Normalized Array: ",normailzed)
```

 27: Sorting a NumPy array based on a specific column

```python
sampleArray = np.array([[34, 43, 73], [82, 22, 12], [53, 94, 66]])


sorted_indices = sampleArray[:, 2].argsort()

sorted_array = sampleArray[sorted_indices]

print("Original array:\n",sampleArray)

print("Sorted Array:\n",sorted_array)
```

28: Delete and Insert a Column in a NumPy Array

```python
sampleArray = np.array([[1, 2, 3], [15, 19, 21], [10, 20, 30]])
new_column = np.array([10, 10, 10])

deletedArray = np.delete(sampleArray, 1, axis=1)

resultedArray = np.insert(deletedArray, 1, new_column, axis= 1)

print(resultedArray)
```

29: Swap column 1 and 2 in a 2D array


```python
# column swap
arr = np.arange(9).reshape(3, 3)
print("Before:\n",arr)

arr[:, [1, 2]] = arr[:, [2, 1]]
print("After:\n",arr)

# row swap
print("Before:\n",arr)

arr[[1, 2], :] = arr[[2, 1], :]
print("After:\n",arr)
```
