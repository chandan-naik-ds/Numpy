##### 1. Create a 1D NumPy array of numbers from 0 to 9

```python
import numpy as np

z = np.arange(10)
print(z)
```

#####  2: Convert 1D array to 2D

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.shape)
arr_2d = arr.reshape(2, 3)
print(arr_2d)
print(arr_2d.shape)
```

##### Exercise 3: Print Array Attributes

```python
arr = np.arange(10)
my_array = np.array([[1, 2], [3, 4], [5, 6], [7, 8]], dtype=np.uint16)
print(my_array.shape)
print(my_array.ndim)
print(my_array.itemsize)
print(arr.shape)
print(arr.ndim)
print(arr.itemsize)
```
