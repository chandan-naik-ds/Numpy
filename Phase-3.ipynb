### total sales per year

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])
print(np.sum(sales_data, axis=0))
yearly_sales = np.sum(sales_data[:, 1:], axis=0)
print(yearly_sales)

# Minimum sales per restaurent
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Minimum sales per year
min_sales = np.min(sales_data[:, 1:], axis=0)
print(min_sales)

# Average sales per restaurent
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)

# Cumulative ales of each restaurent
cumsum = np.cumsum(sales_data[:, 1:],axis=1 )
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average sales acroos the year per restau0rant")
plt.xlabel("Years")
plt.xlabel("Sales")
plt.grid(True)
plt.show()
```
